```python
#xgboost algorithm
import pandas as pd

data = {
"Soil_Type":[
"Loamy","Clayey","Sandy","Loamy","Clayey","Sandy","Loamy","Black","Black","Red",
"Red","Alluvial","Alluvial","Loamy","Clayey","Black","Red","Sandy","Alluvial","Loamy",
"Black","Red","Clayey","Alluvial","Sandy","Black","Red","Loamy","Clayey","Alluvial",
"Loamy","Red","Black","Clayey","Alluvial","Loamy","Red","Sandy","Black","Clayey",
"Alluvial","Loamy","Red","Black","Sandy","Clayey","Alluvial","Loamy","Red","Black"
],


"Rainfall_Level":[
"High","High","Low","Medium","High","Low","Medium","Medium","Low","Medium",
"Low","High","Medium","Medium","High","Medium","Medium","Low","High","Medium",
"Low","Medium","High","Medium","Low","Medium","Medium","Medium","High","High",
"Medium","Low","Medium","High","High","Medium","Medium","Low","Medium","High",
"High","Medium","Medium","Low","Low","High","High","Medium","Medium","Low"
],


"Temperature_Level":[
"Hot","Hot","Hot","Mild","Hot","Hot","Mild","Hot","Hot","Mild",
"Hot","Hot","Mild","Mild","Hot","Hot","Hot","Hot","Hot","Mild",
"Hot","Mild","Hot","Mild","Hot","Hot","Mild","Mild","Hot","Hot",
"Mild","Hot","Hot","Hot","Hot","Mild","Mild","Hot","Hot","Hot",
"Hot","Mild","Mild","Hot","Hot","Hot","Hot","Mild","Mild","Hot"
],


"Soil_pH":[
"Neutral","Acidic","Neutral","Neutral","Acidic","Alkaline","Neutral","Neutral","Alkaline","Neutral",
"Alkaline","Neutral","Neutral","Neutral","Acidic","Neutral","Neutral","Alkaline","Neutral","Neutral",
"Neutral","Neutral","Acidic","Neutral","Alkaline","Neutral","Neutral","Neutral","Acidic","Neutral",
"Neutral","Alkaline","Neutral","Acidic","Neutral","Neutral","Neutral","Alkaline","Neutral","Acidic",
"Neutral","Neutral","Neutral","Neutral","Alkaline","Acidic","Neutral","Neutral","Neutral","Neutral"
],


"Irrigation":[
"Yes","Yes","No","Yes","Yes","No","Yes","Yes","Yes","Yes",
"No","Yes","Yes","Yes","Yes","Yes","Yes","No","Yes","Yes",
"Yes","Yes","Yes","Yes","No","Yes","Yes","Yes","Yes","Yes",
"Yes","No","Yes","Yes","Yes","Yes","Yes","No","Yes","Yes",
"Yes","Yes","Yes","Yes","No","Yes","Yes","Yes","Yes","Yes"
],


"Land_Size":[
"Medium","Large","Small","Medium","Medium","Small","Large","Large","Medium","Medium",
"Small","Large","Medium","Small","Large","Large","Medium","Small","Large","Medium",
"Medium","Small","Large","Medium","Small","Large","Medium","Large","Large","Large",
"Medium","Small","Large","Large","Large","Medium","Small","Small","Medium","Large",
"Large","Medium","Small","Large","Small","Large","Large","Medium","Small","Large"
],


"Season":[
"Kharif","Kharif","Rabi","Rabi","Kharif","Kharif","Rabi","Kharif","Kharif","Rabi",
"Kharif","Kharif","Rabi","Rabi","Kharif","Kharif","Kharif","Kharif","Kharif","Rabi",
"Kharif","Rabi","Kharif","Rabi","Kharif","Kharif","Rabi","Rabi","Kharif","Kharif",
"Rabi","Kharif","Kharif","Kharif","Kharif","Rabi","Rabi","Kharif","Kharif","Kharif",
"Kharif","Rabi","Rabi","Kharif","Kharif","Kharif","Kharif","Rabi","Rabi","Kharif"
],


"Crop_Type":[
"Rice","Rice","Millet","Wheat","Rice","Millet","Wheat","Cotton","Cotton","Groundnut",
"Millet","Sugarcane","Wheat","Barley","Rice","Cotton","Groundnut","Millet","Sugarcane","Wheat",
"Cotton","Groundnut","Rice","Wheat","Millet","Cotton","Groundnut","Wheat","Rice","Sugarcane",
"Wheat","Millet","Cotton","Rice","Sugarcane","Wheat","Groundnut","Millet","Cotton","Rice",
"Sugarcane","Wheat","Groundnut","Cotton","Millet","Rice","Sugarcane","Wheat","Groundnut","Cotton"
]
}
import numpy as np
import pandas as pd
df_xgb = pd.get_dummies(pd.DataFrame(data).drop('Crop_Type', axis=1))
y = np.where(pd.DataFrame(data)['Crop_Type'] == 'Rice', 1, 0).astype(float)
X = df_xgb.values

class SimpleXGBoost:
    def __init__(self, n_estimators=5, learning_rate=0.1):
        self.n_estimators = n_estimators
        self.lr = learning_rate
        self.trees = []
        self.base_pred = None

    def fit(self, X, y):
        self.base_pred = np.mean(y)
        current_preds = np.full(y.shape, self.base_pred)

        for _ in range(self.n_estimators):
            residuals = y - current_preds
            best_feat = 0
            best_split_val = 0
            max_reduction = -float('inf')

            for feat_idx in range(X.shape[1]):
                threshold = 0.5
                left_mask = X[:, feat_idx] <= threshold
                right_mask = X[:, feat_idx] > threshold
                gain = np.var(residuals) - (np.var(residuals[left_mask]) + np.var(residuals[right_mask]))
                if gain > max_reduction:
                    max_reduction = gain
                    best_feat = feat_idx
            left_val = np.mean(residuals[X[:, best_feat] <= 0.5])
            right_val = np.mean(residuals[X[:, best_feat] > 0.5])

            self.trees.append((best_feat, left_val, right_val))
            for i in range(len(X)):
                val = left_val if X[i, best_feat] <= 0.5 else right_val
                current_preds[i] += self.lr * val

    def predict(self, X):
        preds = np.full(X.shape[0], self.base_pred)
        for feat, l_val, r_val in self.trees:
            for i in range(len(X)):
                preds[i] += self.lr * (l_val if X[i, feat] <= 0.5 else r_val)
        return np.where(preds > 0.5, 1, 0)
model = SimpleXGBoost(n_estimators=10, learning_rate=0.3)
model.fit(X, y)
final_preds = model.predict(X)
print(f"XGBoost accuracy for 'Rice': {np.mean(final_preds == y) * 100:.2f}%")
```

```python
#adaboost
import pandas as pd

data = {
"Soil_Type":[
"Loamy","Clayey","Sandy","Loamy","Clayey","Sandy","Loamy","Black","Black","Red",
"Red","Alluvial","Alluvial","Loamy","Clayey","Black","Red","Sandy","Alluvial","Loamy",
"Black","Red","Clayey","Alluvial","Sandy","Black","Red","Loamy","Clayey","Alluvial",
"Loamy","Red","Black","Clayey","Alluvial","Loamy","Red","Sandy","Black","Clayey",
"Alluvial","Loamy","Red","Black","Sandy","Clayey","Alluvial","Loamy","Red","Black"
],


"Rainfall_Level":[
"High","High","Low","Medium","High","Low","Medium","Medium","Low","Medium",
"Low","High","Medium","Medium","High","Medium","Medium","Low","High","Medium",
"Low","Medium","High","Medium","Low","Medium","Medium","Medium","High","High",
"Medium","Low","Medium","High","High","Medium","Medium","Low","Medium","High",
"High","Medium","Medium","Low","Low","High","High","Medium","Medium","Low"
],


"Temperature_Level":[
"Hot","Hot","Hot","Mild","Hot","Hot","Mild","Hot","Hot","Mild",
"Hot","Hot","Mild","Mild","Hot","Hot","Hot","Hot","Hot","Mild",
"Hot","Mild","Hot","Mild","Hot","Hot","Mild","Mild","Hot","Hot",
"Mild","Hot","Hot","Hot","Hot","Mild","Mild","Hot","Hot","Hot",
"Hot","Mild","Mild","Hot","Hot","Hot","Hot","Mild","Mild","Hot"
],


"Soil_pH":[
"Neutral","Acidic","Neutral","Neutral","Acidic","Alkaline","Neutral","Neutral","Alkaline","Neutral",
"Alkaline","Neutral","Neutral","Neutral","Acidic","Neutral","Neutral","Alkaline","Neutral","Neutral",
"Neutral","Neutral","Acidic","Neutral","Alkaline","Neutral","Neutral","Neutral","Acidic","Neutral",
"Neutral","Alkaline","Neutral","Acidic","Neutral","Neutral","Neutral","Alkaline","Neutral","Acidic",
"Neutral","Neutral","Neutral","Neutral","Alkaline","Acidic","Neutral","Neutral","Neutral","Neutral"
],


"Irrigation":[
"Yes","Yes","No","Yes","Yes","No","Yes","Yes","Yes","Yes",
"No","Yes","Yes","Yes","Yes","Yes","Yes","No","Yes","Yes",
"Yes","Yes","Yes","Yes","No","Yes","Yes","Yes","Yes","Yes",
"Yes","No","Yes","Yes","Yes","Yes","Yes","No","Yes","Yes",
"Yes","Yes","Yes","Yes","No","Yes","Yes","Yes","Yes","Yes"
],


"Land_Size":[
"Medium","Large","Small","Medium","Medium","Small","Large","Large","Medium","Medium",
"Small","Large","Medium","Small","Large","Large","Medium","Small","Large","Medium",
"Medium","Small","Large","Medium","Small","Large","Medium","Large","Large","Large",
"Medium","Small","Large","Large","Large","Medium","Small","Small","Medium","Large",
"Large","Medium","Small","Large","Small","Large","Large","Medium","Small","Large"
],


"Season":[
"Kharif","Kharif","Rabi","Rabi","Kharif","Kharif","Rabi","Kharif","Kharif","Rabi",
"Kharif","Kharif","Rabi","Rabi","Kharif","Kharif","Kharif","Kharif","Kharif","Rabi",
"Kharif","Rabi","Kharif","Rabi","Kharif","Kharif","Rabi","Rabi","Kharif","Kharif",
"Rabi","Kharif","Kharif","Kharif","Kharif","Rabi","Rabi","Kharif","Kharif","Kharif",
"Kharif","Rabi","Rabi","Kharif","Kharif","Kharif","Kharif","Rabi","Rabi","Kharif"
],


"Crop_Type":[
"Rice","Rice","Millet","Wheat","Rice","Millet","Wheat","Cotton","Cotton","Groundnut",
"Millet","Sugarcane","Wheat","Barley","Rice","Cotton","Groundnut","Millet","Sugarcane","Wheat",
"Cotton","Groundnut","Rice","Wheat","Millet","Cotton","Groundnut","Wheat","Rice","Sugarcane",
"Wheat","Millet","Cotton","Rice","Sugarcane","Wheat","Groundnut","Millet","Cotton","Rice",
"Sugarcane","Wheat","Groundnut","Cotton","Millet","Rice","Sugarcane","Wheat","Groundnut","Cotton"
]
}
import numpy as np
import pandas as pd
df_ada = pd.get_dummies(pd.DataFrame(data).drop('Crop_Type', axis=1))
X = df_ada.values
y = np.where(pd.DataFrame(data)['Crop_Type'] == 'Rice', 1, -1)

class SimpleAdaBoost:
    def __init__(self, n_stumps=5):
        self.n_stumps = n_stumps
        self.stumps = []
        self.alphas = []

    def fit(self, X, y):
        n_samples, n_features = X.shape
        w = np.full(n_samples, (1 / n_samples))

        for _ in range(self.n_stumps):
            best_stump = {}
            min_error = float('inf')
            for feat_idx in range(n_features):
                threshold = 0.5
                p = 1
                predictions = np.ones(n_samples)
                predictions[X[:, feat_idx] <= threshold] = -1
                error = sum(w[y != predictions])

                if error > 0.5:
                    error = 1 - error
                    p = -1

                if error < min_error:
                    min_error = error
                    best_stump['feat'] = feat_idx
                    best_stump['p'] = p
            eps = 1e-10
            alpha = 0.5 * np.log((1.0 - min_error + eps) / (min_error + eps))
            final_preds = np.ones(n_samples)
            if best_stump['p'] == 1:
                final_preds[X[:, best_stump['feat']] <= 0.5] = -1
            else:
                final_preds[X[:, best_stump['feat']] > 0.5] = -1

            w *= np.exp(-alpha * y * final_preds)
            w /= np.sum(w)

            self.stumps.append(best_stump)
            self.alphas.append(alpha)

    def predict(self, X):
        stump_preds = []
        for i, stump in enumerate(self.stumps):
            predictions = np.ones(X.shape[0])
            if stump['p'] == 1:
                predictions[X[:, stump['feat']] <= 0.5] = -1
            else:
                predictions[X[:, stump['feat']] > 0.5] = -1
            stump_preds.append(self.alphas[i] * predictions)

        return np.sign(np.sum(stump_preds, axis=0))
model = SimpleAdaBoost(n_stumps=10)
model.fit(X, y)
final_preds = model.predict(X)
print(f"AdaBoost Accuracy for 'Rice': {np.mean(final_preds == y) * 100:.2f}%")
```
